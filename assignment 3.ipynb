```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,3,13)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, color = "green")
plt.xlabel("x axis")
plt.ylabel("x axis")
plt.title("this is title")
plt.show()
```

```python
# line plots are used for showcasing the trends , 
```

```python
courses = ["BECE","CSE","BEEE","IT"]
students = [25,65,70,30]
```

```python
plt.bar(courses,students, color = "red")
plt.xlabel("x axis")
plt.ylabel("x axis")
plt.title("this is title")
plt.show()
```

```python
# bar graphs can be used to showcase the count , also for trend , this can be used between two numerical data or one cat data and one numerical data
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("x axis")
plt.title("this is title")
plt.show()
```

```python
plt.scatter(x,y, color = "green")
plt.xlabel("x axis")
plt.ylabel("x axis")
plt.title("this is title")
plt.show()
```

```python
# see the distribution 
```

```python
plt.subplot(2,2,1)#(rows,columns , division number)
plt.plot(x,y , color = "blue")
plt.subplot(2,2,2)#(rows,columns , division number)
plt.bar(x,y , color = "red")
plt.subplot(2,2,3)#(rows,columns , division number)
plt.scatter(x,y , color = "purple")
plt.show()
```

```python
courses = ["BECE","CSE","BEEE","IT"]
students = [25,65,70,30]
plt.pie(students,labels = courses,autopct = "%0.1f%%" ,colors = ["yellow","pink","purple","magenta"])
plt.legend(courses,loc= "upper left")
plt.show()
```

```python
plt.plot(x,y,label = "x vs y")
plt.plot(y,x, label = "y vs x")
plt.legend()
plt.show()
```

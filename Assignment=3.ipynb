```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")#loads the data set iris
```

```python
iris#info about the flowers
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)#scatterplot
```

```python
sns.lineplot(x='sepal_length',y='petal_length',data=iris,ci=None)#lineplot
```

```python
sns.distplot(iris['petal_length'])#distplot
```

```python
sns.barplot(x='species',y='petal_length',data=iris)#barplot
```

```python
sns.countplot(x='petal_width',data=iris)#countplot
```

```python
sns.boxplot(x='sepal_width',y='petal_width',data=iris)#boxplot
```

```python
sns.catplot(x='species',y='petal_length',hue='petal_width',data=iris)#catplot
```

```python
iris.corr()#correlation
```

```python
sns.heatmap(iris.corr(),annot=True)#heatmap
```
